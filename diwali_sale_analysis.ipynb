```python
gender_graph = sns.countplot(x ='Gender', data = df, hue = 'Gender')
for bars in gender_graph.containers:
    gender_graph.bar_label(bars)
plt.title('Distribution of Count and Gender')
plt.show()

#prints gender bar graph with count as y-axis and gender as x-axis
```

```python
sales_generation = df.groupby(['Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = True)
sns.barplot(x = 'Gender', y = 'Amount' , data = sales_generation)
plt.title('Distribution of Amount Spend by Gender')
plt.show()

#sales graph with amount spend on y-axis and gender in x-axis
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(sales_age['Amount'], labels=sales_age['Age Group'], autopct='%1.1f%%', startangle=100, colors=sns.color_palette('Set2'))

plt.legend(wedges, sales_age['Age Group'], title="Age Groups", loc="center left", bbox_to_anchor=(1.1, -0.2, 0.5, 1))
plt.text(1.35, -1.15, 'X-axis: Age Groups')
plt.text(1.35, -1.25, 'Y-axis: Total Amount')
plt.title('Distribution of Total Amount Spent by Age Group')
plt.show()
#rect index box ==>legend contains colors of each age grp
```

```python
top_10_products = df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd',
          '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
fig1, ax1 = plt.subplots(figsize=(12, 7))
top_10_products.plot(kind='bar', color=colors, ax=ax1)

plt.xlabel('Product ID')
plt.ylabel('Number of Orders')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.title('Top 10 Orders and Product ID')
plt.show()
```
